```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importando o arquivo 
senadores = pd.read_csv('C:/Users/eleni/Documents/2019.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')
```

```python
#Analise quantitativa dos dados
senadores.describe()
```

```python
#Total de reembolso
senadores['VALOR_REEMBOLSADO'].sum() 
```

```python
#Reebolsos por cada senador
senadores['SENADOR'].value_counts()
```

```python
#Quanto cada senador pediu de reembolso
senadores.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
#5 maiores valores de reembolso
senadores.nlargest(5, 'VALOR_REEMBOLSADO')
```

```python
#5 menores valores de reembolso
senadores.nsmallest(5, 'VALOR_REEMBOLSADO')
```

```python
#Quantidade de reembolso por tipo de dispesa
senadores['TIPO_DESPESA'].value_counts()
```

```python
# Quais são os valores totais de reembolso por tipo de despesa?
senadores.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```
